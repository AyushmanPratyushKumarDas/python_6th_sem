```python
from textblob import TextBlob

text = "The quick brown fox jumps over the lazy dog. The sun shines brightly in the clear blue sky. Birds are singing sweet melodies in the morning air. A gentle breeze rustles through the green leaves. Children play happily in the nearby park."

blob = TextBlob(text)
print(blob)
```

```python
print(blob.sentences)
print(blob.words)
```

```python
import nltk
nltk.download('averaged_perceptron_tagger') #for tags
nltk.download('brown') ##for noun_phrases
```

```python
print(blob.tags)
print(blob.noun_phrases)
```

```python
print(blob.sentiment)
print(blob.sentiment.polarity)
print(blob.sentiment.subjectivity)
```

```python
for sentence in blob.sentences:
    print(sentence.sentiment)
```

## some more imp


```python
from pathlib import Path
text = Path('Beneath_the_Crimson_Veil.txt').read_text(encoding='utf-8')
blob=TextBlob(text)
```

```python
items = blob.word_counts.items()
items
```

```python
[item for item in items if item[0]]
```

```python
## sorting the words by frequency
from operator import itemgetter
sorted_items = sorted(items,key = itemgetter(1),reverse=True)
sorted_items
```

```python
top20 = sorted_items[0:21]
top20
```

```python
import pandas as pd
df = pd.DataFrame(top20,columns=['word','count'])
df
```

```python
axes = df.plot.bar(x='word',y='count',legend = True)
axes
```

```python
from pathlib import Path
import imageio.v2 as imageio
from wordcloud import WordCloud
import matplotlib.pyplot as plt

text = Path('Beneath_the_Crimson_Veil.txt').read_text('utf-8')
mask_img = imageio.imread('heart.png')

wordcloud = WordCloud(colormap='prism',mask = mask_img,background_color='white')
wordcloud = wordcloud.generate(text)

plt.figure(figsize=(8,8))
plt.imshow(wordcloud,interpolation = 'bilinear')
plt.axis('off')
plt.show()
```
